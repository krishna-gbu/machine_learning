## matplotlib example workflow

```python
# 0.import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x=[1,2,3,4]
y=[11,22,33,55]
#2. setup plot
fig,ax = plt.subplots(figsize=(5,5))
#3.plot data
ax.plot(x,y);
#4. customize plot
ax.set(title='simple plot',xlabel='x-axis',ylabel='y-axis')
#5. save & show (you save the whole figure)
fig.savefig('image/sample.png')
```

## two  options for subplots

```python
# subplot option 1 
# hum utni hi varible create krenge jitni row and colom me element hoga
#(ax1,ax2) = one row
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x);
ax2.scatter(np.sin(x/2),np.sin(x));
ax3.bar(nut_butter.values(),nut_butter.keys())
ax4.hist(x);
```

```python
# subplot option 2
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.sin(x),np.sin(x/2));
ax[1,0].bar(x,x/2);
ax[1,1].hist(x);
```

### WHICH one should use ? (pyplot vs matplotlib OO method?)
* when plotting something quickly use pyplot
* when plotting something more advance , use the OO method

```python
heart_dise
```

```python
over_50 =heart_dise[heart_dise['age']>50]
over_50.head()
```

```python
# pyplot method
over_50.plot(x='age',y='chol',c='target',kind='scatter');
```

```python
# OO method mixed with pyplot method // object oriented method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax);

# limit of x axis 45 to 100 length
# ax.set_xlim([45,100]) 
```

```python
## oo method from scratch
fig,ax = plt.subplots(figsize=(10,6));
# plot the data 
scatter = ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target']);
# customize the plot
ax.set(title ='heart disease and alcohal level',xlabel='Age',ylabel='Cholestrol');
# add a legend 
ax.legend(*scatter.legend_elements(),title='target');
# add a horizontal level
ax.axhline(over_50['chol'].mean(),linestyle='--');
```

```python
over_50.head()
```

```python
# subplot of chol , age , thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1, figsize=(10,10))
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])
ax0.set(title='heart disease and chalostrol levels',xlabel='age',ylabel='cholestrol');
ax0.axhline(y=over_50['chol'].mean(),linestyle='--');

# ax1
ax1.scatter(x=over_50['age'],
           y=over_50['thalach'],
           c=over_50['target'])
ax1.set(title='heart diseases and  max heart',xlabel='age',ylabel='thalach');
ax1.axhline(y=over_50['thalach'].mean(),linestyle='--');
fig.suptitle('heart diseses');
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')
## oo method from scratch
fig,ax = plt.subplots(figsize=(10,6));
# plot the data 
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                    cmap='winter'); ##color change
# customize the plot
ax.set(title ='heart disease and alcohal level',xlabel='Age',ylabel='Cholestrol');
# add a legend 
ax.legend(*scatter.legend_elements(),title='target');
# add a horizontal level
ax.axhline(over_50['chol'].mean(),linestyle='--');
```

```python
#customizing the y and x axis limitations

# subplot of chol , age , thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1, figsize=(10,10))
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter')
ax0.set(title='heart disease and chalostrol levels',xlabel='age',ylabel='cholestrol'
       );
# change the limit of x axis
ax0.set_xlim([50,80])
ax0.axhline(y=over_50['chol'].mean(),linestyle='--');

# ax1
ax1.scatter(x=over_50['age'],
           y=over_50['thalach'],
           c=over_50['target'])
ax1.set(title='heart diseases and  max heart',xlabel='age',ylabel='thalach');
ax1.axhline(y=over_50['thalach'].mean(),linestyle='--');
ax1.set_xlim([50,80])
fig.suptitle('heart diseses');
```
